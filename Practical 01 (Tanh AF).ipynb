```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(temp_arr):
    return np.tanh(temp_arr)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function : Tanh')
plt.show()
print("\nName : Shantanu Anant Gaikwad\nRoll No : 27")
```
